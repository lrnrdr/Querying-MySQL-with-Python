```python
# PyMySQL is tool that allows us to connect to our MySQL database from Python. 
# We will import it and then install it as our MySQL database.
import pymysql
pymysql.install_as_MySQLdb()
```

```python
# We will also use SQLAlchemy. You will need to include the following imports:
from sqlalchemy import create_engine
import pandas as pd
```

```python
# Create connection string using credentials following this format
# connection = "dialect+driver://username:password@host:port/database"
username = "root"
password = "root" # (whatever password you chose during mysql installation)
db_name = "world"
connection = f"mysql+pymysql://root:root@localhost/world"
```

```python
# Create the engine. To make the connection between Python and MySQL, 
# you will create an engine using the connection string you just defined called connection.

engine = create_engine(connection)
```

```python
# To confirm that it worked, type "engine" and verify the output:
engine
```

```python
# create connection to database via the engine
conn = engine.connect()
```

```python
# Execute Query. You can also execute any SQL query with pd.read_sql. 
# Remember, you can write multi-line strings by using triple quotes around them.

q = """SELECT * FROM countries;"""
pd.read_sql(q, conn)
```

```python
# Queries Using LIKE % .   You may sometimes get an error when
# executing a query that uses the LIKE keyword contains a "%".
# Example Error with a query that contains LIKE and  %
q = """SELECT name FROM countries
WHERE name LIKE 'An%';"""
pd.read_sql(q, conn)
```

```python
# To get around this problem, import the "text" function from sqlalchemy
# and pass the query into the text function. See the example below:

# Import the text function
from sqlalchemy import text
q = """SELECT name FROM countries
WHERE name LIKE 'An%';"""
# Pass the query though the text function before running read_sql
pd.read_sql(text(q), conn)
```
